# Basic Operations with Dataframes

```python
# Mathematical Operations: 
results = stats.BirthRate * stats.InternetUsers
```

```python
results.head()
```

```python
# Add Column
stats['MyCalc'] = stats.BirthRate * stats.InternetUsers
stats.head()
```

```python
# Comparison to R
# stats['xyz'] = [1,2,3,4,5] # Error in Python since row counts do not match. No recycling option
```

```python
# Removing Column
stats = stats.drop('MyCalc', axis=1)
stats.head()
```

# Filtering Data Frames

```python
stats.InternetUsers < 2
```

```python
filter = stats.InternetUsers < 2
```

```python
stats[filter] # Conceptually just like R 
```

```python
stats.BirthRate > 40
```

```python
filter2 = stats.BirthRate > 40
stats[filter2]
```

```python
# Error using 'and' since Python is not a vectorized language. 'and' only works with single values
filter and filter2
```

```python
# For 'and': use &, *
stats[filter & filter2]
```

```python
# For 'or': use |, +
stats[filter + filter2]
```

```python
stats[stats.IncomeGroup == 'High income']
```

```python
# Uniques
stats.IncomeGroup.unique()
```

```python
# Exercise: Find out everything about Malta
stats[stats.CountryName.str.lower().str.contains('^malta')]
```

# Intro to Seaborn

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
plt.rcParams['figure.figsize'] = 8, 4
sns.set_style('darkgrid')

import warnings 
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
# Distribution: 
vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis1 = sns.distplot(stats['InternetUsers'], bins=30)
```

```python
# Boxplots:
vis2 = sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

# Seaborn Gallery
[https://seaborn.pydata.org/](https://seaborn.pydata.org/index.html)

```python
# Linear Model Plot
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False)
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup', size=10, aspect=1)
```

# Keyword Arguments in Python (advanced)

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=False, hue='IncomeGroup', \
    size=10, aspect=1, scatter_kws={"s": 100})
# scatter_kws refers to pyplot's scatter
```
