```python
x = list(range(5))
print(
    x,
    type(x),
    sep='\n'
)
```

```python
# lists can have mixed data types
y = [4, 'string', True, 0.5]
y.append('45')
y.append(list(range(4)))
print(
    y,
    y.index(4),
    y[0],
    sep='\n'
)
```

```python
z = list(range(1, 8))
w = list(range(5, 130, 5))
print(
    z,
    w,
    sep='\n'
)
```

```python
# python has two indexations for lists
# starts at 0 from left
# starts at -1 from right
print(
    x[0],
    len(y),
    y[0:-1:2],
    y[1:-1:3],
    sep='\n'
)
```

```python
import time as t
```

```python
l1 = list(range(15))
print(
    l1[2:len(l1):3],
    l1[:],
    t.time(),
    l1[::-1],
    t.time(),
    l1[7::-1],
    sep='\n'
)
```
